# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
print(tips.head(5))
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(tips.shape)
```

### Проверьте есть ли пропуски в данных

```python
print(tips.isna().sum())
```

### Посмотрите на распределение числовых признаков

```python
print(tips.describe())
```

### Найдите максимальное значение 'total_bill'

```python
max_bill = tips['total_bill'].max()
print(max_bill)
```

### Найдите количество курящих людей

```python
smoker_count = tips['smoker'].value_counts()['Yes']
print(smoker_count)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
mean_bill_by_day = tips.groupby('day', observed=False)['total_bill'].mean()
print(mean_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_bill = tips['total_bill'].median()
high_bill_tips = tips[tips['total_bill'] > median_bill]
mean_tip_by_sex = high_bill_tips.groupby('sex', observed=False)['tip'].mean()
print(mean_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], kde=True, color='skyblue')
plt.title('Распределение общего счета (Total Bill)')
plt.xlabel('Общий счет')
plt.ylabel('Частота')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill', y='tip', data=tips, color='purple')
plt.title('Взаимосвязь между общим счетом и чаевыми')
plt.xlabel('Общий счет')
plt.ylabel('Чаевые')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2', hue='day', legend=False)
plt.title('Распределение общего счета по дням недели')
plt.xlabel('День недели')
plt.ylabel('Общий счет')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
g = sns.FacetGrid(tips, col='time', height=5)
g.map(sns.histplot, 'tip', kde=True, color='salmon')
g.set_axis_labels('Чаевые', 'Частота')
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
g_scatter = sns.FacetGrid(tips, col='sex', hue='smoker', height=5, palette='Set1')
g_scatter.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.7)
g_scatter.add_legend(title='Курильщик (1-Да, 0-Нет)')
plt.show()
```
